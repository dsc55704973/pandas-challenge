```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data["SN"].unique())
player_count
df = pd.DataFrame({player_count})
total_players_df = df.rename(columns={0:"Total Players"})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# compute unique items using Item ID
unique_items = len(purchase_data["Item ID"].unique())
# compute number of purchases using Price and len() method
purchase_volume = len(purchase_data["Price"])
# compute Total Revenue using Price and sum() method
total_revenue = purchase_data["Price"].sum()
total_revenue_dollars = f"${total_revenue}"
# compute average price using Price
average_price = round(total_revenue / purchase_volume, 2)
average_price_dollars = f"${average_price}"
# create final dataframe
summary_df = pd.DataFrame({
    "Unique Items": unique_items,
    "Average Price": average_price_dollars,
    "Purchase Volume": [purchase_volume],
    "Total Revenue": total_revenue_dollars})
summary_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#retrieve Item ID, Item Name, and Item Price columns
purchase_data.head()
df = purchase_data.loc[: ,["Item ID","Item Name","Price"]]
df

# groupby Item ID, grab 
grouped_df = df.groupby(["Item ID"])
purchase_count = grouped_df.count()
purchase_count
purchase_count = purchase_count.loc[:, "Item Name"] # grab Item Name data

# add Purchase Count and Total Purchase Value to dataframe
df["Purchase Count"] = purchase_count # add Purchase Count
indexed_df = df.set_index("Item ID") # set index to Item ID
# add Total Purchase Value
indexed_df["Total Purchase Value"] = indexed_df["Price"] * indexed_df["Purchase Count"]
# sort by Total Purchase Price in descending order
sorted_df = indexed_df.sort_values("Total Purchase Value", ascending=False)
sorted_df
```
